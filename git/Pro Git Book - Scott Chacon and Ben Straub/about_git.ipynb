## Data Chain

- Git hashes are computed based on the changes you make PLUS the hash of the previous commit, creating a chain that is prohibitively expensive to "break"

- It's somewhat like blockchain, in that you have cryptographic hashes that are logically tied to each other and cannot be modified without changing the entire history, which is super expensive
